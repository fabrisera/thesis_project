```python
import pandas as pd
import os
```

```python
# for each file in directory, read the stata file and print the column names
predictor_dictionary = {}
for file in os.listdir("../../data/CDP Cleaning/cleaned outputs"):
    if file.endswith(".dta"):
        df = pd.read_stata("../../data/CDP Cleaning/cleaned outputs/" + file)
        # count the number of nan for each column
        
        # get data dypes of each column
        predictor_dictionary[file] = list(zip(df.columns, df.dtypes, df.isna().mean()))
```

```python
predictor_dictionary
```

```python
df_list = []
for df in predictor_dictionary.values():
    df_list.append(pd.DataFrame(predictor_dictionary['cdp_carbon_credits_voluntary_purchase.dta'], columns = ['variable', 'dtype', 'nan_count']))
```

```python
# concat the dfs and use a dobule index to identify the file name
df = pd.concat(df_list, keys = predictor_dictionary.keys())
```

```python
# print the df to a csv including the double index
df.to_csv("../../data/processed/variable_dictionary.csv", index = True)
```
