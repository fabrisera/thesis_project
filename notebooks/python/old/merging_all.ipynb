```python
directory_path = '../../data/CDP Cleaning/cleaned outputs'
rows = {}
# Read all files and clean them
for file in os.listdir(directory_path):
    # if file ends in .dta read it and store name and number of rows
    if file.endswith(".dta"):
        df = pd.read_stata(os.path.join(directory_path, file))
        rows[file] = df.shape[0]
```

```python
# sort by number of rows
sorted_rows = sorted(rows.items(), key=lambda x: x[1], reverse=True)
sorted_rows
```
